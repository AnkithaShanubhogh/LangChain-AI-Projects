```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

# LangSmith (optional)
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = "QnA Chatbot Notebook"
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Respond clearly."),
    ("user", "Question: {question}")
])
```

```python
def get_llm(provider, api_key, model, temperature):
    if provider == "openrouter":
        return ChatOpenAI(
            model=model,
            temperature=temperature,
            base_url="https://openrouter.ai/api/v1",
            api_key=api_key
        )
    elif provider == "groq":
        return ChatGroq(
            model=model,
            temperature=temperature,
            groq_api_key=api_key
        )
```

```python
# ==== USER INPUT SECTION ====
provider = input("Choose provider (openrouter/groq): ")
api_key = input("Enter API Key: ")
model = input("Enter model name: ")
temperature = float(input("Temperature (0-1): ") or 0.7)

question = input("Ask your question: ")

# ==== RUN CHAIN ====
llm = get_llm(provider, api_key, model, temperature)
output_parser = StrOutputParser()

chain = prompt | llm | output_parser
response = chain.invoke({"question": question})

print("\nResponse:\n")
print(response)
```
